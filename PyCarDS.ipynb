# Pearson Correlation
->Correlation coefficient
  ->Close to +1:Large Positive Realtionship
  ->Close to -1:Large Negative Relationship
  ->Close to  0:No Relationship
->P-Value
  ->Pvalue < 0.001 Strong Certainity in result
  ->Pvalue < 0.05  Moderate Certainity in result
  ->Pvalue < 0.1   Weak Certainity in result
  ->Pvalue > 0.1   No Certainity in result
->Strong Correlation: 
    ->Correlation coefficient close to 1 or -1
    ->P value < 0.001

```python
lm=LinearRegression()
lm
```

#define predictor variable and target variable

```python
X=df2[['highway-mpg','curb-weight','engine-size','horsepower']]
Y=df2['price']
```

#use lm.fit(X,Y) to fit model,i.e. fine the parameters b0 and b1 

```python
lm.fit(X,Y)
```

#Obtain prediction using method prediction()

```python
Yhat=lm.predict(X)
Yhat[0:5]
```

```python
a=lm.intercept_
```

#Slope/coefficient/parameter of x:(b1)

```python
b=lm.coef_
```

```python
Yhat=a+(b*X)
Yhat
```
